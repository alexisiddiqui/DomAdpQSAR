```python
### Notebook to test DNN Class|
from DomAdpQSAR.utility import gpu, seed_all

from DomAdpQSAR.QSARsettings import Settings
```

```python
from DomAdpQSAR.QSARdnn import DomAdpQSARDNN
test_settings = Settings()
test_settings.trial_name = "Basic Tuning"
test_Experiment = DomAdpQSARDNN(test_settings)

test_Experiment.load_featuriser()

test_Experiment.settings.epochs_to_run = 200
```

```python
featuriser_names =[
    "Broad Tuning Underfit", 
    "Broad Tuning Underfit Ranked",
    "Broad Tuning Overfit",
    "Gradual Fine Tuning 15 Epochs",
    "Gradual Sub Sampling 15 Epochs",
    "Gradual Fine Tuning GFT Rank 15 Epochs",
]


for featuriser_name in featuriser_names:
    featuriser_path = "logs/" + featuriser_name + "/"
    test_Experiment.load_featuriser(featuriser_path)

    test_Experiment.settings.trial_name = featuriser_name+ " SS TF Clean Tuning"
    test_Experiment.dataset_setup()
    test_Experiment.model_setup()
    test_Experiment.set_rank(None)
    test_Experiment.set_training_data(test_Experiment.clean_dataset_loader)
    test_Experiment.train()

    test_Experiment.settings.trial_name = featuriser_name+" SS TF Clean Tuning Ranked"
    test_Experiment.dataset_setup()
    test_Experiment.model_setup()
    test_Experiment.set_rank('Tanimoto')
    test_Experiment.set_training_data(test_Experiment.clean_dataset_loader)
    test_Experiment.train()
```

```python
test_Experiment.settings.trial_name = "SS TF Clean Tuning"
test_Experiment.dataset_setup()
test_Experiment.model_setup()
test_Experiment.set_training_data(test_Experiment.clean_dataset_loader)
test_Experiment.train()
```
